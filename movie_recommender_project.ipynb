```python
#genres
#id
#keywards
#title
#overview
#cast
#crew
movies=movies[['movie_id', 'title','overview','genres','keywords','cast','crew']]
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
import ast

def convert_genre(text):
    if pd.isna(text):
        return []
    return [i['name'] for i in ast.literal_eval(text)]

movies['genres'] = movies['genres'].apply(convert_genre)
```

```python
import ast

def fetch_names(obj):
    return [i['name'] for i in ast.literal_eval(obj)]


movies['keywords'] = movies['keywords'].apply(fetch_names)
```

```python
import ast

def fetch_cast_str(obj):
    # Case 1: already a list (broken or clean)
    if isinstance(obj, list):
        # join characters back, then split by space
        names = "".join(obj).split()
        return " ".join(names[:3])

    # Case 2: string (original TMDB format)
    cast = ast.literal_eval(obj)
    return " ".join([i['name'].replace(" ", "") for i in cast[:3]])
movies['cast'] = movies['cast'].apply(fetch_cast_str)

```

```python
import ast

def fetch_director(obj):
    crew = ast.literal_eval(obj)
    for i in crew:
        if i['job'] == 'Director':
            return [i['name']]
    return []
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
import re

movies['overview'] = movies['overview'].apply(
    lambda x: re.sub(r'[^a-zA-Z0-9 ]', '', x).split() if isinstance(x, str) else x
)


```

```python
movies['genres']   = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])

movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])

movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = (
    movies['overview'] +
    movies['genres'] +
    movies['keywords'] +
    movies['cast'] +
    movies['crew']
)
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
from nltk.tokenize import word_tokenize
import nltk
from nltk.stem.porter import PorterStemmer

# Setup
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
ps = PorterStemmer()

# Helper function
def string_to_stemmed_list(text):
    words = word_tokenize(text.lower())
    stemmed_words = [ps.stem(word) for word in words]
    return stemmed_words  # returns the list

# Example usage
text = "This is a simple testing example"
result = string_to_stemmed_list(text)  # store in a variable
print(result)  # optional, just to check

```

```python
new_df['tags']=new_df['tags'].apply(string_to_stemmed_list)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
 movie_index=new_df[new_df['title'] == movie].index[0]
 distances=similarity[movie_index]
 movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
 for i in movies_list:
  print(new_df.iloc[i[0]].title)
 return
```
